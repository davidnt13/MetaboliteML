```python
import pandas as pd
from rdkit import Chem
from tqdm.auto import tqdm
from rdkit.Chem.Draw import MolsToGridImage
import seaborn as sns
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors.MoleculeDescriptors import MolecularDescriptorCalculator
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

origData = pd.read_csv("allDescriptors.csv")

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
allData = imp_mean.fit_transform(origData)
allData = pd.DataFrame(data = allData, columns = origData.columns)

allTargets = pd.read_csv("allTargets.csv")
```

```python
from sklearn.preprocessing import StandardScaler
allDataVals = StandardScaler().fit_transform(allData.values)
allTargetVals = allTargets.values
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 50)
allDataComp = pca.fit_transform(allDataVals)
```

```python
tsne = TSNE(n_components = 2)
tsne_crds = tsne.fit_transform(allDataVals)
```

```python
tsne_df = pd.DataFrame(tsne_crds,columns=["X","Y"])
list1 = [0] * 4724
list2 = [1] * 66
vals = list1 + list2
tsne_df['is_metabolite'] = vals
tsne_df.head()
```

```python
ax = sns.scatterplot(data=tsne_df.query("is_metabolite == 0"),x="X",y="Y",color='lightblue', label = "Non-Metabolite")
ax = sns.scatterplot(data=tsne_df.query("is_metabolite == 1"),x="X",y="Y",color='red', label = "Metabolite")
plt.legend()
```
